This the error analysis we will have at the target feature and compare the ML results (y-predicted) to the actual values (y-actual)

At first we will need to load the orginal data after preprocessing. 
Then we will load the the y-predicted data and plot y-predicted vs y-actual. 



```python
#improting needed labiraries 
import pandas as pd #for df ops 
import numpy as np # for array ops 
import matplotlib.pyplot as plt #for plotting
import seaborn as sns #for plotting

from preprocessing import get_preprocessor #loading preprocessors pipe 
from sklearn import set_config #no clue why
set_config(transform_output="pandas")

from sklearn.metrics import r2_score # loading metrics for evaluation later on 
from sklearn.metrics import mean_squared_error

from sklearn.pipeline import Pipeline #loading pipline package so we can use the preprocessing pipeline
```

```python
#loading actual train and test data 
df_test             = pd.read_csv('df_test.csv') #maybe not needed
df_train            = pd.read_csv('df_train.csv')

#loading actual y data in log from 
y_train_log         = pd.read_csv('y_train_log.csv') 
y_train             = np.exp(y_train_log)
y_test_log          = pd.read_csv('y_test_log.csv') 
y_test             = np.exp(y_test_log)



#loading y-predicted values from test data
y_hat_test_elastic  = pd.read_csv('y_hat_test_elastic.csv')
y_hat_test_lin_reg  = pd.read_csv('y_hat_test_lin_reg.csv')
y_hat_test_lin_reg = np.exp(y_hat_test_lin_reg)
y_hat_test_RForest  = pd.read_csv('y_hat_test_RForest.csv')
y_hat_test_STree    = pd.read_csv('y_hat_test_STree.csv')
#loading y-predicted values from train data
y_hat_train_elastic = pd.read_csv('y_hat_train_elastic.csv') #maybe not needed
y_hat_train_lin_reg = pd.read_csv('y_hat_train_lin_reg.csv')
y_hat_train_lin_reg = np.exp(y_hat_train_lin_reg)

y_hat_train_RForest = pd.read_csv('y_hat_train_RForest.csv')
y_hat_train_STree   = pd.read_csv('y_hat_train_STree.csv')
```

Next we will create a plot to display y vs y predicted 

```python

sns.regplot(data=None, x=y_test, y=y_hat_test_lin_reg, line_kws={"color": "green"})
plt.plot([min(y_test.values), max(y_test.values)], [min(y_hat_test_lin_reg.values), max(y_hat_test_lin_reg.values)], 'r')
plt.legend(['data', 'regressionline', None, 'perfect world' ])

plt.xlabel('actual')
plt.ylabel('predicted')

# Calculate the R-squared for our model
print("R-squared train:", r2_score(y_test, y_hat_test_lin_reg).round(3))
# # mean squared error
print("Root mean squared error test:", np.sqrt(mean_squared_error(y_test, y_hat_test_lin_reg).round(3)))
```
