```python
import pandas as pd
import numpy as np
from preprocessing import get_preprocessor
from sklearn.model_selection import train_test_split
from sklearn import set_config
set_config(transform_output="pandas")
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor  
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv("https://storage.googleapis.com/second-try-ml/air-pollution/Train.csv")
df.head()
```

```python
RSEED = 42
X = df.drop("target", axis=1)
y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, shuffle=True, random_state=RSEED)
df_train = pd.concat([X_train, y_train], axis=1)
df_train.head(5)
```

```python
y_train_log = y_train.apply(lambda x: np.log(x + 1))
y_test_log = y_test.apply(lambda x: np.log(x + 1))
```

```python
preproc = get_preprocessor()
X_train_pp = preproc.fit_transform(X_train)
X_test_pp = preproc.transform(X_test)
```

```python
lin_reg = Pipeline([
    ('Preprocessing', preproc),
    ('LinearRegression', LinearRegression())
]) 
```

```python

# Train the model using our data
lin_reg.fit(X_train, y_train_log)

# Calculate the estimated value for each data point
y_hat = lin_reg.predict(X_test)

y_hat_train = lin_reg.predict(X_train)

# Calculate the R-squared for our model
print("R-squared:", r2_score(y_test_log, y_hat).round(3))
print("R-squared train:", r2_score(y_train_log, y_hat_train).round(3))

# # mean squared error
print("Mean squared error:", mean_squared_error(y_test_log, y_hat).round(3))
print("Mean squared error train:", mean_squared_error(y_train_log, y_hat_train).round(3))
```

```python
from sklearn.linear_model import ElasticNet
elastic = Pipeline([
    ('Preprocessing', preproc),
    ('LinearRegression', ElasticNet())
])


# Train the model using our data
elastic.fit(X_train, y_train)

# Calculate the estimated value for each data point
y_hat = elastic.predict(X_test)

y_hat_train = elastic.predict(X_train)


# Calculate the R-squared for our model
print("R-squared:", r2_score(y_test, y_hat).round(3))
print("R-squared train:", r2_score(y_train, y_hat_train).round(3))

# # mean squared error
print("Mean squared error:", mean_squared_error(y_test, y_hat).round(3))
print("Mean squared error train:", mean_squared_error(y_train, y_hat_train).round(3))
```

```python
# import the regressor 
singleTree = Pipeline([
    ('Preprocessing', preproc),
    ('Tree', DecisionTreeRegressor(max_depth=10))
])

  
# fit the regressor with X and Y data 
singleTree.fit(X_train, y_train) 

y_hat = singleTree.predict(X_test)
y_hat_train = singleTree.predict(X_train)


# Calculate the R-squared for our model
print("R-squared:", r2_score(y_test, y_hat).round(3))
print("R-squared train:", r2_score(y_train, y_hat_train).round(3))

# # mean squared error
print("Mean squared error:", mean_squared_error(y_test, y_hat).round(3))
print("Mean squared error train:", mean_squared_error(y_train, y_hat_train).round(3))
```

```python

randomForest = Pipeline([
    ('Preprocessing', preproc),
    ('RandomForest', RandomForestRegressor(max_depth=70, bootstrap=True))
])

# fit the regressor with X and Y data 
randomForest.fit(X_train, y_train) 

y_hat = randomForest.predict(X_test)
y_hat_train = randomForest.predict(X_train)


# Calculate the R-squared for our model
print("R-squared:", r2_score(y_test, y_hat).round(3))
print("R-squared train:", r2_score(y_train, y_hat_train).round(3))

# # mean squared error
print("Mean squared error:", mean_squared_error(y_test, y_hat).round(3))
print("Mean squared error train:", mean_squared_error(y_train, y_hat_train).round(3))
```

```python
# Define parameter grid for randomized search

len()
param_grid = {"max_samples" : [2,4,3,5,10], #samples for bootstrap per tree
              "weights" : ["uniform", "distance"],
              "p" : [1, 2, 3],
              "algorithm": ["ball_tree", "kd_tree", "brute"]
             }
```
